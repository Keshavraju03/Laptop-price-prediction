```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import math
```

```python
company = []
processer = []
ram = []
os = []
ssd = []
hdd = []
ssize = []
Ms = []
price = []
ignore=[]
for i in range(1,41):
    data = requests.get(f'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page={i}')
    soup = BeautifulSoup(data.content,'html')
    # Extracting the product(laptop) name
    a = soup.find_all('div',attrs={
        'class':'_4rR01T'
    })
    for i in a:
        a = i.text
        b = a.split()
        c = b[0]
        company.append(c)
    d = soup.find_all('li',attrs={
        'class':'rgWa7D'
    })
    for i in d:
        e = i.text
        f = e.split()
        if 'RAM' and 'SSD' and 'Storage' in f:
            ignore.append(f)
        elif 'Display' and 'nits' in f:
            ignore.append(f)
        elif f.count('Processor') == 1:
            os.append(f[0])
            processer.append(f[1]+' '+f[2])
        elif f.count('RAM') == 1 and f.count('GB'):
            ram.append(f[0])
        elif f.count('Display') == 1 and f.count('cm') == 1:
            ssize.append(f[0])
            
    g = soup.find_all('li',attrs={
        'class':'rgWa7D'
    })
    for i in g:
        h = i.text
        j = h.split()
        if 'RAM' and 'SSD'and 'Storage' in j:
            ignore.append(j)
        elif j.count('HDD') == 0 and j.count('SSD')==1:
            ssd.append(j[j.index('SSD')-2]+j[j.index('SSD')-1])
            hdd.append(0)
        elif j.count('HDD') == 1 and j.count('SSD')==1:
            hdd.append(j[j.index('HDD')-2]+j[j.index('HDD')-1])
            ssd.append(j[j.index('SSD')-2]+j[j.index('SSD')-1])
        elif j.count('HDD') == 1 and j.count('SDD')==0:
            hdd.append(j[j.index('HDD')-2]+j[j.index('HDD')-1])
            ssd.append(0)
                       
    n = soup.find_all('div',attrs={
        'class':'fMghEO'
    })
    for i in n:
        o = i.text
        if 'Office' in o:
            Ms.append('Yes')
        else:
            Ms.append('No')
    k = soup.find_all('div',attrs={
        'class':'_30jeq3 _1_WHN1'
    })
    for i in k:
        q = i.text
        r = q.replace('â‚¹', '')
        s = q.replace(',','')
        price.append(s)
    

```

```python
pl = ['Athlon Dual','Celeron Dual','Celeron Quad','Core i3','Core i5','Core i7','Core i9','M1 Max','M1 Pro','M1 Processor'
     ,'M2 Max','M2 Pro','M2 Processor','MediaTek Kompanio','MediaTek MT8788','Pentium Quad','Pentium Silver','Ryzen 3',
     'Ryzen 5','Ryzen 7','Ryzen 9']
processer_m = []
for i in processer:
    if i in pl:
        processer_m.append(i)   
```

```python
ram_m = []
for i in ram:
    if (i.isdigit()):
        ram_m.append(int(i))   
```

```python
os_m = []
standard = ['Intel','AMD','Apple','MediaTek']
for i in os:
    if i in standard:
        os_m.append(i)
```

```python
hdd_m = []
for i in hdd:
    if i=='1TB':
        hdd_m.append(int(1024))
    elif i=='500GB':
        hdd_m.append(int(516))
    else:
        hdd_m.append(0)
```

```python
ssd_m = []
for i in ssd:
    i = str(i)
    if '128GB' in i:
        ssd_m.append(int(128))
    elif '256GB' in i:
        ssd_m.append(int(256))
    elif '512GB' in i:
        ssd_m.append(int(512))
    elif '1TB' in i:
        ssd_m.append(int(1024))
    elif '2TB' in i:
        ssd_m.append(int(2000))
    elif '4TB' in i:
        ssd_m.append(int(4000))
    
        
```

```python
price_m = []
for i in price:
    z = i[1:]
    price_m.append(int(z))
```

```python
ran = 900
company = company[:ran]
processer_m = processer_m[:ran]
ram_m = ram_m[:ran]
os_m = os_m[:ran]
hdd_m = hdd_m[:ran]
ssd_m = ssd_m[:ran]
ssize_m = ssize[:ran]
Ms = Ms[:ran]
price_m = price_m[:ran]
```

```python
def size(p):
    i = float(p)
    inpt = math.ceil(i)
    if inpt>28 and inpt<=33:
        return 12
    elif inpt>33 and inpt<=37:
        return 14
    elif inpt>37 and inpt<=40:
        return 15
    elif inpt>40 and inpt<=45 :
        return 17
    else:
        return 24
```

```python
df['ScreenSize'] = df['ScreenSize'].apply(size)
```

# Model Selection and Building

```python
import sklearn
```

```python
X = data.drop('Price', axis=1)
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

## Hyper parameter tunning

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_

best_model.score(X_test, y_test)
```
